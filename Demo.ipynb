```python
from matplotlib import pyplot as plt
import numpy as np
import math
from pylab import *
```

```python
x = np.arange(0, math.pi*2, 2)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
```

```python
x = np.arange(0, math.pi*2, 0.125)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
#plt.show() Ovde nije neophodno jer je takvo okruzenje, npr. u VS code mora show
```

```python
x = np.arange(0, math.pi*2, 0.125)
y = np.sin(x)
plt.plot(x,y, "g")
#plt.plot(x, y, "r--")
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
#plt.show() Ovde nije neophodno jer je takvo okruzenje, npr. u VS code mora show
```

```python
x = np.arange(0, math.pi*2, 0.125)
y = np.sin(x)
x1 = np.arange(0, math.pi*2, 0.125)
y1 = np.cos(x1)
plt.plot(x1,y1, "g:", linewidth = 3)
plt.plot(x, y, "r--", linewidth = 4)
plt.xlabel("angle")
plt.ylabel("sine and cosine")
plt.title('sine and cosine wave')
#plt.show() Ovde nije neophodno jer je takvo okruzenje, npr. u VS code mora show
```
