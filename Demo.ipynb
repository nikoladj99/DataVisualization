```python
from matplotlib import pyplot as plt
import numpy as np
import math
from pylab import *
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0, 0.2, 0.5, 0.5])
ax.plot(x,y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')

fig.set_size_inches(8, 6)  # Postavljanje dimenzija figure na 8x6 ina
fig.set_facecolor('lightgray')  # Postavljanje boje pozadine figure na svetlo sivu
fig.set_edgecolor('black')  # Postavljanje boje granica figure na crnu

fig.savefig('slika.png')  # uvanje figure u slikovnom formatu kao "slika.png"

plt.show()
```

```python
fig = plt.figure()

# Prvi podgrafik
ax1 = fig.add_axes([0.1, 0.1, 0.4, 0.8])  # [left, bottom, width, height]
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3], 'r')  # Crtanje linije u prvom podgrafiku
ax1.set_xlabel('X oznaka 1')
ax1.set_ylabel('Y oznaka 1')
ax1.set_title('Prvi podgrafik')

# Drugi podgrafik
ax2 = fig.add_axes([0.8, 0.1, 0.4, 0.8])  # [left, bottom, width, height]
ax2.plot([1, 2, 3, 4], [4, 2, 3, 1], 'g')  # Crtanje linije u drugom podgrafiku
ax2.set_xlabel('X oznaka 2')
ax2.set_ylabel('Y oznaka 2')
ax2.set_title('Drugi podgrafik')

# Prikazivanje figure
plt.show()
```

```python
# Kreiranje prvog objekta Figure
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax1.set_xlabel('X label 1')
ax1.set_ylabel('Y label 1')
ax1.set_title('Figure 1')

# Kreiranje drugog objekta Figure
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.plot([1, 2, 3, 4], [4, 2, 3, 1])
ax2.set_xlabel('X label 2')
ax2.set_ylabel('Y label 2')
ax2.set_title('Figure 2')

# Prikazivanje prvog objekta Figure
plt.show(fig1)

# Prikazivanje drugog objekta Figure
plt.show(fig2)

#plt.show() prikazuje sve figure
```

```python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
x3 = [1,2,3,18,28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green colour and circle marker
l3 = ax.plot(x3,y,'r') # dash line with green colour and circle marker
ax.legend(labels = ('tv', 'Smartphone', "dodatno"), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```

```python
fig,a =  plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)
x = np.arange(1,10)
a2.plot(x, x*x)
a2.set_title('square')
a1.plot(x, np.exp(x))
a1.set_title('exp')
a3.plot(x, np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25)
axes[1].set_title('custom grid')
axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x = np.arange(1,5)
axes[0].plot( x, np.exp(x))
axes[0].plot(x,x**2)
axes[0].set_title("Normal scale")
axes[1].plot (x, np.exp(x))
axes[1].plot(x, x**2)
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")
axes[0].xaxis.labelpad = 10
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel('angle')
ax.set_title('sine')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
plt.show()
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```
